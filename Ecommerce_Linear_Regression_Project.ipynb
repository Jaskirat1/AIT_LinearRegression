```python
X = df[['Avg. Session Length', 'Time on App', 'Time on Website', 'Length of Membership']]
y = df['Yearly Amount Spent']
```

```python
X
```

```python
y
```

 **Trian-Test Split**

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state =101)
```

Training the model

```python
from sklearn.linear_model import LinearRegression
```

Create an instance of LinearRegression()model named -lr

```python
lr = LinearRegression()
```

Train/fir lr on the training data

```python
lr.fit(X_train, y_train)
```

```python
print(lr.coef_)
```

Predicting test data

```python
prd = lr.predict(X_test)
```

Create a scatterplot of the real test values versus the predicted values.

```python
plt.scatter(y_test,prd)
```

```python
import sklearn.metrics as metrics
print('MAE: {}'.format(metrics.mean_absolute_error(y_test, prd)))
print('MSE: {}'.format(metrics.mean_squared_error(y_test, prd)))
print('RMSE: {}'.format(np.sqrt(metrics.mean_squared_error(y_test, prd))))
```

```python
sns.distplot((y_test,prd))
```
