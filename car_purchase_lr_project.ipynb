```python
X = df[['Gender', 'Age', 'Annual Salary', 'Net Worth', 'Credit Card Debt']]
y = df['Car Purchase Amount']
```

Train/Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=101) 
```

Training the model 

```python
from sklearn.linear_model import LinearRegression
```

Create an instance of LinearRegression()model 

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
print(lr.coef_)
```

```python
prd = lr.predict(X_test)
```

```python
plt.scatter(y_test, prd)
```

```python
import sklearn.metrics as metrics
```

```python
MAE = metrics.mean_absolute_error(y_test, prd)
print("Mean Absolute Error:=", MAE)
```

```python
sns.distplot((y_test, prd))
```
